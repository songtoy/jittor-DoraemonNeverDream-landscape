```python
import os
from PIL import Image

name = 'standard-with-data-aug'
#name = 'style_vae_best'
stage = 'latest'
path = './results/' + name + '/test_' + stage + '/images/synthesized_image'
targetpath = path.replace('synthesized_image', 'synthesized_image_scaled')
def covert_dir(dirpath):
    faillist = []
    for root, dirs, files in os.walk(dirpath):
        for file in files:
            img = Image.open(os.path.join(dirpath, file))
            img = img.resize((512, 384), Image.BICUBIC)
            img.save(os.path.join(targetpath, file))
            print(f"Processing {file}")

def convert_file(filepath):
    newfilepath = filepath.replace('torch', 'jt')
    with open(filepath, "r") as f:
        code = f.read()
    with open(newfilepath, "w") as f:
        f.write(code)
    print(f"Success at {newfilepath}")
    return True

print(path)
covert_dir(path)
```
